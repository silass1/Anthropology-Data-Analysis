```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from pylab import text

pd.set_option('display.max_columns', None)
plt.rcParams['axes.facecolor'] = 'whitesmoke'
plt.rcParams["figure.figsize"]=(20,20)
plt.rcParams["figure.facecolor"]='white'

df_iq = pd.read_excel("/Users/silas/Documents/College/Spring 2022/Anthro Homework Research/Homework 3/IQ.xlsx")
#df_iq
usa = gpd.read_file("/Users/silas/Documents/College/Spring 2022/Anthro Homework Research/Homework 3/states_21basic/states.shp")
plt.rcParams["figure.figsize"]=(10,10)
not_main = ["Alaska", "Hawaii"]
main = usa.query('STATE_NAME not in @not_main')
main.plot()
```

```python
merge = pd.merge(main, df_iq, on='STATE_NAME')
merge
```

```python
plt.rcParams["figure.figsize"]=(20,20)
merge.plot(column='IQ', cmap='Blues', legend=True, legend_kwds={'shrink':.4})
plt.xticks(ticks=[])
plt.yticks(ticks=[])
plt.title(("Average IQ by State"), fontsize=25, fontweight='bold', pad=15)
plt.text(-125, 27, 'Highest IQ: Massachusetts, 104.3', fontsize=15)
plt.text(-125, 25, 'Lowest IQ: Mississippi, 94.2', fontsize=15)
```
