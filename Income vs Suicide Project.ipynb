```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle 
plt.rcParams["figure.figsize"]=(10,10)

plt.rcParams['axes.facecolor'] = 'lightgrey'

df = pd.read_excel("/Users/silas/Documents/College/Spring 2022/" "Anthro Homework Research/income_unemp_suicide.xlsx", skiprows=3)
df = df.replace(',','', regex=True) 
#df['mean'] = df.iloc[:, [2,6,10,14,18].mean(axis=1)
df
```

```python
bins = [35000, 40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000, 80000, 85000, 90000]
plt.xticks(ticks=bins)
plt.title("Income Distribution: 2014-2018", fontsize=20, fontweight="bold", pad=20)
plt.xlabel("Median Income ($)", fontsize=15, labelpad=10)
plt.ylabel("Count of States", fontsize=15, labelpad=10)
hist_1 = plt.hist((df.income_1, df.income_2, df.income_3, df.income_4, df.income_5), bins=bins, edgecolor="black")
plt.legend(['2014', '2015', '2016', '2017', '2018'])

```

```python
i1 = df.income_1
i2 = df.income_2
i3 = df.income_3
i4 = df.income_4
i5 = df.income_5
s1 = df.suicide_1
s2 = df.suicide_2
s3 = df.suicide_3
s4 = df.suicide_4
s5 = df.suicide_5
u1 = df.unemp_1
u2 = df.unemp_2
u3 = df.unemp_3
u4 = df.unemp_4
u5 = df.unemp_5

i_all = [i1, i2, i3, i4, i5]
i_concat = pd.concat(i_all, ignore_index=True)
s_all = [s1, s2, s3, s4, s5]
s_concat = pd.concat(s_all, ignore_index=True)
u_all = [u1, u2, u3, u4, u5]
u_concat = pd.concat(u_all, ignore_index=True)

u_sc1 = plt.scatter(x=i1, y=u1, color="r")
u_sc2 = plt.scatter(x=i2, y=u2, color="b")
u_sc3 = plt.scatter(x=i3, y=u3, color="g")
u_sc4 = plt.scatter(x=i4, y=u4, color="orange")
u_sc5 = plt.scatter(x=i5, y=u5, color="magenta")

plt.legend((u_sc1, u_sc2, u_sc3, u_sc4, u_sc5),
          ('2014', '2015', '2016', '2017', '2018'),
          scatterpoints=1,
          loc='lower left',
          ncol=1,
          fontsize=8)
plt.plot(np.unique(i_concat), np.poly1d(np.polyfit(i_concat, u_concat, 1))(np.unique(i_concat)), color="black")
plt.title('Median Income VS Unemployment Rate', fontsize=20, fontweight="bold", pad=20)
plt.xlabel(('Median Income ($)'), fontsize=15, labelpad=10)
plt.ylabel(('Unemployment (%)'), fontsize=15, labelpad=10)

plt.yticks(ticks=[2.0,2.5,3.0,3.5,4.0,4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0])

plt.show()
```

```python
i_avg = plt.scatter(x=df.avg_income, y=df.avg_unemp, color="b")
plt.title('Median Income VS Unemployment: Average', fontsize=20, fontweight="bold", pad=20)
plt.xlabel('Median Income ($)', fontsize=15, labelpad=10)
plt.ylabel('Unemployment Rate (%)', fontsize=15, labelpad=10)

state_col = df['State']
income_col = df['avg_income']
unemp_col = df['avg_unemp']

text = state_col.to_list()
x = income_col.to_list()
y = unemp_col.to_list()

for i in range(len(text)):
    if(text[i] == 'FL'):
        plt.annotate(text[i], (x[i] - 1000, y[i] - .05))
    elif(text[i] == 'OR'):
        plt.annotate(text[i], (x[i] + 300, y[i] - .05))
    elif(text[i] == 'TN'):
        plt.annotate(text[i], (x[i] - 1100, y[i] - .05))
    elif(text[i] == 'TX'):
        plt.annotate(text[i], (x[i] - 500, y[i] - .1))  
    else:
        plt.annotate(text[i], (x[i], y[i] + .05))

plt.plot(np.unique(df.avg_income), np.poly1d(np.polyfit(df.avg_income, df.avg_unemp, 1))(np.unique(df.avg_income)), linewidth=2, color="black", alpha=.3)

plt.show()
```

```python
s_sc1 = plt.scatter(x=i1, y=s1, color="r")
s_sc2 = plt.scatter(x=i2, y=s2, color="b")
s_sc3 = plt.scatter(x=i3, y=s3, color="g")
s_sc4 = plt.scatter(x=i4, y=s4, color="orange")
s_sc5 = plt.scatter(x=i5, y=s5, color="magenta")

plt.legend((s_sc1, s_sc2, s_sc3, s_sc4, s_sc5),
          ('2014', '2015', '2016', '2017', '2018'),
          scatterpoints=1,
          loc='lower left',
          ncol=1,
          fontsize=8)
plt.plot(np.unique(i_concat), np.poly1d(np.polyfit(i_concat, s_concat, 1))(np.unique(i_concat)), color="black")
plt.title('Median Income VS Suicide Rate', fontsize=20, fontweight="bold", pad=20)
plt.xlabel(('Median Income ($)'), fontsize=15, labelpad=10)
plt.ylabel(('Suicide Rate (per 100,000 people)'), fontsize=15, labelpad=10)
plt.yticks(ticks=[7.5,10.0,12.5,15.0,17.5,20.0,22.5,25.0,25.0,27.5])
plt.show()
```
